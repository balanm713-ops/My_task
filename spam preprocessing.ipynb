Removing unwanted columns 

```python
madhu = madhu.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'])
madhu
```

Rename the Column

```python
madhu = madhu.rename(columns={'v1': 'label', 'v2': 'message'})
madhu
```

```python
madhu = madhu.dropna()
```

Encoding Label

```python
encoder = LabelEncoder()
madhu['label'] = encoder.fit_transform(madhu['label'])
madhu.head()
```

Text Cleaning 

```python
import re

def clean_text(text):
    text = text.lower()                           # lowercase
    text = re.sub(r'[^a-z\s]', '', text)          # remove numbers & punctuation
    text = re.sub(r'\s+', ' ', text).strip()      # remove extra spaces
    return text

madhu['clean_message'] = madhu['message'].apply(clean_text)
madhu
```

Remove Stopwords

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

madhu['clean_message'] = madhu['clean_message'].apply(
    lambda x: ' '.join([word for word in x.split() if word not in stop_words])
)
madhu.head()
```

Stemmer

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()


madhu['stemmed_message'] = madhu['clean_message'].apply(
    lambda x: ' '.join([ps.stem(word) for word in x.split()])
)
madhu[['clean_message', 'stemmed_message']].head()
```

Lemmatization

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet

# Download required NLTK data
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')



lm = WordNetLemmatizer()

def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {
        'J': wordnet.ADJ,
        'N': wordnet.NOUN,
        'V': wordnet.VERB,
        'R': wordnet.ADV
    }
    return tag_dict.get(tag, wordnet.NOUN)

# Apply Lemmatization
madhu['lemmatized_message'] = madhu['clean_message'].apply(
    lambda x: ' '.join([lm.lemmatize(word, get_wordnet_pos(word)) for word in x.split()])
)


madhu[[ 'stemmed_message', 'lemmatized_message']].head(10)
```
