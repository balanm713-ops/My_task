```python
# Numerical method :
# Histogram...
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

num = pd.read_csv('C:\\Users\XNovaQ.9\\Downloads\\archive (5)\\AirQuality.csv',sep=';')
plt.hist(num['CO(GT)'].dropna(), bins=20, color="black", edgecolor="red")

plt.xlabel('CO ')
plt.ylabel('Frequency')
plt.title(' CO Levels')
```

```python
# Categorical method :
# Bar chart...
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

cate = pd.read_csv('C:\\Users\\XNovaQ.9\\Downloads\\archive (6)\\penguins.csv')
countspecies = cate['species'].value_counts()
plt.bar(countspecies.index,countspecies.values,color = 'red',edgecolor = 'black')
plt.xlabel('Species')
plt.ylabel('Count')

plt.show()
```
