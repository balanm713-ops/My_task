```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from scipy.stats import chi2_contingency

ObserevredValues = np.array([[23,24],[12,15]])

rows = ["Male","Female"]
columns = ["Offline","Online"]

chi , p , dof , excepted = chi2_contingency(ObserevredValues)

formula = ((ObserevredValues - excepted )**2)/excepted

print("Observed values : ",ObserevredValues)
print(excepted.round(3))
print(formula.round(4))
print("Degree of Freedom :",dof)
print(p)

alpha = 0.05
if p < alpha:
    print("\nReject H0 â†’ Significant relationship between variables.")
else:
    print("\nFail to reject H0 â†’ No significant relationship between variables.")


df_observed = pd.DataFrame(ObserevredValues, index=rows, columns=columns)
df_expected = pd.DataFrame(excepted, index=rows, columns=columns)

df_observed.plot(kind='bar')
plt.title("Observed Frequncies")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Class Preference")
plt.show()


df_observed.plot(kind='bar', stacked=True)
plt.title("Observed Frequencies (Stacked)")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Class Preference")
plt.show()


plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.heatmap(df_observed, annot=True, cmap="Blues", fmt="g")
plt.title("Observed Frequencies")

plt.subplot(1,2,2)
sns.heatmap(df_expected, annot=True, cmap="Oranges", fmt=".2f")
plt.title("Expected Frequencies")

plt.tight_layout()
plt.show()
```
